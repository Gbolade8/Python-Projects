```python
from google.colab import files
import pandas as pd

# Upload the file
uploaded = files.upload()

# Load it into a DataFrame
df = pd.read_csv(next(iter(uploaded)))

# Verify the data
df.head()
```

```python
df.columns
```

```python
print(df.columns)  # Display all column names
```

```python
# Check structure and missing values
df.info()

# Summary statistics for numeric columns
df.describe()

# Display first few rows
df.head()
```

```python
# Churn distribution
df['Churn'].value_counts().plot(kind='bar', title='Churn Distribution')

# Churn by Contract Type
df.groupby('Contract')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True, title='Churn by Contract Type')
```

```python
# Internet Service distribution
df['Internet Service'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Internet Service Types')

# Churn rate by Internet Service
df.groupby('Internet Service')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True, title='Churn by Internet Service')
```

```python
# Payment method distribution
df['Payment Method'].value_counts().plot(kind='bar', title='Payment Method Distribution')

# Churn rate by Payment Method
df.groupby('Payment Method')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', stacked=True, title='Churn by Payment Method')
```

```python
# Streaming TV and Movies
df[['Streaming TV', 'Streaming Movies']].apply(pd.Series.value_counts).plot(kind='bar', title='Streaming Services Subscription')

# Churn rate by Streaming TV
df.groupby('Streaming TV')['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', title='Churn by Streaming TV')
```
